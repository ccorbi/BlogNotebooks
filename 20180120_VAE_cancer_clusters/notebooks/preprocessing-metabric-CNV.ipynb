```python
def intersect(*d):
    sets = iter(map(set, d))
    result = sets.next()
    for s in sets:
        result = result.intersection(s)
    return result
```

```python
def transformation(row):

        if row == -2:
            return 0
        if row == -1:
            return .25
        if row == 0:
            return .5
        if row == 1:
            return .75
        if row == 2:
            return 1
        else:
            print(row)
```

```python
for c in meta_cnv_reorga.columns:
    if c == 'Hugo_Symbol':
        pass
    else:
        meta_cnv_reorga[c] = meta_cnv_reorga[c].apply(transformation)
```
